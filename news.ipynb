```python
import requests as req
from bs4 import BeautifulSoup
import re

url = 'https://sports.news.naver.com/kfootball/news/index.nhn?isphoto=N'

res = req.get(url)
text = res.text
print(text)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
title = soup.select('.title span')
print(title)
```

```python
title = re.sub('<.+?>', '', str(title) ,0)
title = re.sub(',','\n', str(title), 0)

print(title)
```
